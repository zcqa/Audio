```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
import librosa.display
```

```python
duke_file = 'duke.wav'
ipd.Audio(duke_file)
y, sr = librosa.load(duke_file)
```

## MFCC

```python
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
print(mfccs.shape)
```

## 可视化

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.show()
```

## first/second MFCCS derivatives

```python
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs, x_axis='time')
plt.colorbar()
plt.title('Delta MFCC')
plt.show()

plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs, x_axis='time')
plt.colorbar()
plt.title('Delta2 MFCC')
plt.show()
```

```python
#拼接
mfccs_features = np.concatenate((mfccs, delta_mfccs, delta2_mfccs), axis=0)
```
