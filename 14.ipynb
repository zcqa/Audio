```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
import librosa.display
```

```python
piano_file = "piano_c.wav"
ipd.Audio(piano_file)  # Listen to the audio
piano_c5,sr = librosa.load(piano_file)

```

```python
plt.figure(figsize=(15, 17))

plt.subplot(3, 1, 1)
librosa.display.waveshow(piano_c5, alpha=0.5)
plt.ylim((-1, 1))
plt.title("piano_c5")

plt.show()
```

```python
def plot__magnitude_spectrum(signal,sr,title,f_ratio=1):
    X = np.fft.fft(signal)
    X_mag = np.abs(X)

    plt.figure(figsize=(18, 5))
    f = np.linspace(0, sr, len(X))
    f_bins = int(len(X) * f_ratio)

    plt.plot(f[:f_bins], X_mag[:f_bins])
    plt.xlabel('Frequency (Hz)')
    plt.title(title)
```

```python
plot__magnitude_spectrum(piano_c5,sr,"piano_c5",f_ratio=0.1)
```
